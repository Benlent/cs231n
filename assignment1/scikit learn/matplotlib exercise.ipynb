```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
n = 256
X = np.linspace(-np.pi, np.pi, n, endpoint=True)
Y = np.sin(2 * X)

plt.axes([0.025, 0.025, 0.95, 0.95])

plt.plot(X, Y + 1, color='blue', alpha=1.00, linewidth = 0.8)
plt.plot(X, Y - 1, color='blue', alpha=1.00, linewidth = 0.8)


L = np.ones((256, ))
plt.plot(X, L, color = 'none')
plt.plot(X, -L, color = 'none')


plt.fill_between(X, Y + 1, 1, color = 'blue', alpha = 0.25)

plt.fill_between(X, Y - 1, -1, (Y - 1) < -1, color = 'red', alpha = .25)
plt.fill_between(X, Y - 1, -1, (Y - 1) > -1, color = 'blue', alpha = .25)

plt.xlim(-np.pi, np.pi)
plt.ylim(-2.5, 2.5)

plt.xticks(())
plt.yticks(())

```

```python
fig = plt.figure()
fig.subplots_adjust(bottom=0.025, left=0.025, top = 0.975, right=0.975)

plt.subplot(2, 1, 1)
plt.xticks(())
plt.yticks(())

plt.subplot(2, 3, 4)
plt.xticks(())
plt.yticks(())

plt.subplot(2, 3, 5)
plt.xticks(())
plt.yticks(())

plt.subplot(2, 3, 6)
plt.xticks(())
plt.yticks(())
```
